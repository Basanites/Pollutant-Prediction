```python
import pandas as pd
import matplotlib.pyplot as plt
from fbprophet import Prophet
import os
%matplotlib inline

package_dir = os.getcwd()

df = pd.read_csv(os.path.join(package_dir, "res/DE_1_6085_2013_timeseries.csv"),
                 encoding="utf-16", parse_dates=[13, 14], 
                 infer_datetime_format=True, 
                 index_col=[4, 13])

df = df.sort_index()

print(df.dtypes)

```

```python
dfu = df.unstack(level=0)
dfu.plot(y='Concentration')
```

```python
dfu = dfu[dfu.Validity == 1]
dfu.plot(y='Concentration')
```
